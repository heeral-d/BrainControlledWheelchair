````
# Initial imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv
Read the data and display the first 5 rows

````
#Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv

file_url = "http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv"
titanic_data = pd.read_csv(file_url)
titanic_data.head()
````

### Preprocessing of the Dataset



---


**1. Convert the categorical values of columns 'sex', 'embarked', 'pclass' into dummy values and add them into titanic_data_new DataFrame**




---



````
titanic_data_new = titanic_data.filter(['sex','embarked','pclass'], axis=1)
titanic_data_new.head()

#sex = pd.get_dummies(dataframe['sex'])
#embarked = pd.get_dummies(dataframe['embarked'])
#pclass = pd.get_dummies(dataframe['pclass'])
#titanic_data_new = pd.concat([sex, embarked, pclass], axis = 1)
````



---


**2. Fill the missing data in Age and Fare columns with the mean value of the respected column then add these columns to titanic_data_new**


---



````
age = pd.DataFrame(titanic_data['age']) 
#age.isnull()
age_new = age.fillna(age.mean())
#print(age_new)

fare = pd.DataFrame(titanic_data['fare'])
fare_new = fare.fillna(fare.mean())

titanic_data_new['age'] = age_new
titanic_data_new['fare'] = fare_new
print(titanic_data_new)
````

###Normalization


---



1. Normalize the columns which have numerical data



---



````
titanic_num = titanic_data.select_dtypes(include=[np.number])
titanic_norm = (titanic_num - titanic_num.mean()) / (titanic_num.max() - titanic_num.min())
#titanic_data[titanic_norm.columns] = titanic_norm
titanic_norm.head()
````


---
2. Drop the original age, fare columns
3. Append the normalized columns to the titanic_data_new DataFrame.



---



````
titanic_data_new = titanic_data_new.drop(['age', 'fare'], axis = 1)
titanic_data_new.head()
````

````
titanic_data_new['age'] = titanic_norm['age']
titanic_data_new['fare'] = titanic_norm['fare']
titanic_data_new.head()
````

###Plotting
